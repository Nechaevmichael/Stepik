```python
# Импорт модуля math  и работа с ним
import math as m
import pprint
print(m.pi)
print(m.e)
print(m.inf)
print(m.nan)
print(m.tau)
print(m.log(8, 2))
print(m.pow(2, 3))
print(m.factorial(5))
pprint.pprint(locals())
```
